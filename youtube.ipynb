```python
# Read CSV into dataframe 
df = pd.read_csv(r"C:\Users\USER\Downloads\data analysis project\youtube_data_germany.csv")

```

```python
# Extract channel IDs and remove potential duplicates
df['channel_id'] = df['NAME'].str.split('@', n=1).str[-1].str.strip()
channel_ids = df['channel_id'].unique()
```

```python
import time
import random

def get_channel_stats(youtube, channel_id, retries=3):
    for attempt in range(retries):
        try:
            request = youtube.channels().list(
                part='snippet,statistics',
                id=channel_id
            )
            response = request.execute()
            
            if 'items' in response and response['items']:
                item = response['items'][0]
                subs = int(item['statistics'].get('subscriberCount', 0))
                views = int(item['statistics'].get('viewCount', 0))
                vids = int(item['statistics'].get('videoCount', 0))
                return {
                    'channel_name': item['snippet']['title'],
                    'total_subscribers': subs,
                    'total_views': views,
                    'total_videos': vids
                }
            else:
                print(f"No data returned for channel {channel_id}")
                return None
            
        except Exception as e:
            print(f"Attempt {attempt+1} failed for {channel_id}: {e}")
            wait = (2 ** attempt) + random.random()  # exponential backoff
            time.sleep(wait)
    print(f"Failed to fetch data for {channel_id} after {retries} attempts.")
    return None
```

```python


#channel_id = df['NAME'].str.split('@').str[-1].unique()
```

```python
# Initialize a list to keep track of channel stats
channel_stats = []
```

```python
# Loop over the channel IDs and get stats for each
for channel_id in channel_ids:
    stats = get_channel_stats(youtube, channel_id)
    if stats is not None:
        channel_stats.append(stats)
```

```python
# Convert the list of stats to a df
stats_df = pd.DataFrame(channel_stats)
```

```python
df.reset_index(drop=True, inplace=True)
stats_df.reset_index(drop=True, inplace=True)
```

```python
# Concatenate the dataframes horizontally
combined_df = pd.concat([df, stats_df], axis=1)
```
